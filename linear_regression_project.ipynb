```python
from sklearn.datasets import load_diabetes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

source = load_diabetes()
df = pd.DataFrame(source["data"], columns=["age","sex","bmi","map", "tc", "ldl", "hdl","tch", "ltg", "glu"])
df["index"] = source["target"]

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,2], df["index"], test_size=0.1)
X_train = np.array(X_train).reshape(-1,1)
X_test = np.array(X_test).reshape(-1,1)
model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)
print(f"Mean Square Error: {mean_squared_error(y_test, predictions)}")

plt.scatter(X_test, y_test, color = "blue")
plt.plot(X_test, predictions, color = "black")
plt.title("BMI vs Diabetes Index")
plt.xlabel("BMI")
plt.ylabel("Index")
_ = plt

# Using other independent variables does not give a significant trend related to diabetes.
```
